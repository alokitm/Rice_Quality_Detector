```python
# ============================================
# Import Libraries
# ============================================

import os
from ultralytics import YOLO
```

```python
# ============================================
# Dataset Path
# ============================================

dataset_path = "/kaggle/input/datasets/alokit1301/rice-quality-dataset"

print(os.listdir(dataset_path))
```

```python
# ============================================
# Check Dataset Structure
# ============================================

print(os.listdir(dataset_path + "/train"))
print(os.listdir(dataset_path + "/valid"))
print(os.listdir(dataset_path + "/test"))
```

```python
# ============================================
# Check YAML File
# ============================================

yaml_path = dataset_path + "/data.yaml"

print(yaml_path)
```

```python
# ============================================
# Load YOLOv8 Model
# ============================================

# yolov8n.pt = nano model (fast)

model = YOLO("yolov8n.pt")
```

```python
print(model) # for checking yolo is loaded or not 
```

```python
# ============================================
# Train YOLOv8 Model
# ============================================

results = model.train(

    data=yaml_path,   # dataset yaml file

    epochs=50,        # training epochs

    imgsz=640,        # image size

    batch=8,          # batch size

    device='cpu',

    project="Rice_Detection_Project",

    name="YOLOv8_Rice_Model"
)
```

```python
# ============================================
# Validate Model
# ============================================

metrics = model.val()

print(metrics)
```

```python
# ============================================
# Predict on Test Images
# ============================================

results = model.predict(

    source=dataset_path + "/test/images",

    conf=0.25,

    save=True
)
```

```python
# ============================================
# Rice Detection + Counting
# ============================================

from ultralytics import YOLO
from collections import Counter

# Load trained YOLO model
model = YOLO(
    "/kaggle/working/runs/detect/Rice_Detection_Project/YOLOv8_Rice_Model/weights/best.pt"
)

# Predict on test images
results = model.predict(

    source="/kaggle/input/datasets/alokit1301/rice-quality-dataset/test/images",

    conf=0.25,

    save=True
)

# Class names
class_names = model.names

# Store detected classes
detected_classes = []

# Loop through predictions
for result in results:

    boxes = result.boxes

    for box in boxes:

        # Class ID
        class_id = int(box.cls[0])

        # Convert class ID to class name
        class_name = class_names[class_id]

        detected_classes.append(class_name)

# Count each class
counts = Counter(detected_classes)

# Print counts
print("Rice Counts:\n")

for cls, count in counts.items():

    print(f"{cls} : {count}")
```
